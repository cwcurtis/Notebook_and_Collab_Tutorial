```python
# Core libraries you will need for computing and plotting.  DO NOT TOUCH!!!!!!!!!!!!!!!!!!

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

PI = np.pi
```

```python
# Core functions you will need for computing and plotting.  DO NOT TOUCH!!!!!!!!!!!!!!!!!!

def xline(a,b):
    return np.linspace(a,b,np.int(1e3)+1)

def fun_plot(a,b,sz,*args):
    xvals = xline(a,b)
    plt.figure(figsize=(sz, sz))
    cnt = 0
    for arg in args:
        fvals = arg(xvals)
        labels = r'$f_{%d}(\theta)$' % cnt
        plt.plot(xvals,fvals,label=labels)
        cnt += 1
    plt.xlabel(r'$\theta$', fontsize=24)
    plt.legend(loc='best', fontsize=12)
```

## Trigonometry In Action

In the following, we are going to see how different trigonometric functions compare to one another, and how weirdly for something that is defined via triangles, we end up spending so much time going round and round a circle as implied in the following figure 

<img src="https://upload.wikimedia.org/wikipedia/commons/b/b5/Sin-cos-defn-I.png" style="height: 300px; width:300px;"/>

To wit, first let us wrap our heads around the idea of a _phase shift_.  To do so, we are going to look at functions of the form

$$
f(\theta) = \sin(\theta - \phi), ~ 0\leq \theta \leq 2\pi, ~ 0 \leq \phi \leq 2\pi.
$$

We call $\phi$ the phase shift.  So we could try to explain what this means in words, but honestly it is better to just see it in action.  To wit, we are going to look at five different functions given by 

\begin{align}
f_{0}(\theta) = & \sin\left(\theta\right)\\
f_{1}(\theta) = & \sin\left(\theta - \frac{\pi}{4}\right)\\
f_{2}(\theta) = & \sin\left(\theta - \frac{\pi}{2}\right)\\
f_{3}(\theta) = & \sin\left(\theta - \pi\right)
\end{align}

Note, we say that $f_{1}(\theta)$ is _phase-shifted_ by $\pi/4$ radians relative to $f_{0}(\theta)$, $f_{2}(\theta)$ is phase-shifted by $\pi/2$ radians relative to $f_{0}(\theta)$ and so forth.  But what does this mean when we plot each function over the interval $0\leq \theta \leq 2\pi$?  Well first, we need to write a little bit of code.  

```python
a = 0. # Left endpoint of the interval you want to plot.  
b = 2.*PI # Right endpoint of the interval you want to plot.
sz = 10. # Window size of figure.  
fun0 = lambda x: np.sin(x)
fun1 = lambda x: np.sin(x - PI/4.)
fun2 = lambda x: np.sin(x - PI/2.)
fun3 = lambda x: np.sin(x - PI)
```
